## Examples

**Example 1**

A program to verify age and check Country to see if elligible to German Exchange program

```python
user_age = int(input('What is your age? '))
user_country = input('What is your country? ')

if user_age < 25 and user_country == 'Germany':
    print('You can apply for a German student exchange programme')
else:
    print('Sorry, you do not qualify')
```

**Example 2**

Multiple conditions on age

```python
user_age = int(input('What is your age? '))
if user_age > 30:
    print('You are over 30 years old')
    print('Sorry, you do not qualify')
elif user_age == 30:
    print('You are exactly 30 years old')
    print('You will need to meet additional conditions to qualify')
else:
    print('You are 30 years old or younger')
    print('Congratulations, you qualify!')
```

**Example 3**

Multi conditions with **or**

```python
user_country = input('What is your country?')

if user_country == 'Sweden' or user_country == 'Denmark' or user_country == 'Norway':
    print('You can apply for a Scandinavian student exchange programme')
else:
    print('Sorry, you do not qualify')
```

**Example 4**

```python
user_age = int(input('What is your age? '))
user_country = input('What is your country? ')

if ((not user_country == 'Germany') and user_age) < 25 or \
    (user_country == 'Germany' and user_age > 30):
    print('You qualify!')
else:
    print('You don\'t qualify!')
```

## II. While loops <a id="1"></a>

<code>while</code> statement in permforms iteration. 

A <code>while</code> statement will repeatedly execute a single statement or group of statements as long as the condition is True. 

<code>while</code> loops  over and over again until the condition is no longer met.

Feneral format of a while loop is:

    while test:
        code statements
    else:
        final code statements

```python
x = 0
while x < 6:
    print('x is : ',str(x) + '___still inferior to 6')
    x+=1
```

```python
for x in range(1000):
    print('x is:',x)
    if x == 6:
        print('Now x = 6')
        break
```

**Add an else condition**

```python
x = 0
while x < 6:
    print('x is : ',str(x) + '___still inferior to 6')
    x+=1
else:
    print('Condition is met and x is:',x)
```

**Add a continue condition**

```python
x = 0
while x < 6:
    print('x is : ',str(x) + '___still inferior to 6')
    x+=1
    if x == 2:
        print('Nothing is actually done here, just continuing with same logic')
        continue
else:
    print('Condition is met and x is:',x)
```
